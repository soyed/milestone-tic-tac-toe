```python
import random 
# Create global variables
game_board = [' ' for num in range(9)]
player_tokens = {"player1": "", "player2": ""}
game_starter = {"player1": False, "player2": False}
game_score = {"player1": 0, "player2": 0}
token_counter = {"player1": 0, "player2": 0}
random_num = random.randint(0,10)
current_player = 'player1'
# Winning combinations
winning_comb = ["012","345","678","048","246","036","147","258"]
```

```python
# Create a welcome message
def welcome_message() -> None:
    print("Welcome! This is a simple tic tac toe game!")
    print("The rules are simple: ")
    print("\t The first player to have matching tokens either vertically, horiontally or diagonally wins!")
    print("\t The game is drawn if all spots on the board are filled and there is not match")
    print("\t To make the the game fair:")
    print("\t\t Both players must guess a number between 0 to 10 (inclusive of both bounds),")
    print("\t\t If any player guess the right number they get to start!")
    print("\t\t The player whose guess falls between a range of +/- 3 of the random number starts the game!")
    print("Enjoy!!\n")
```

```python
# Create board function || Display board function
def display_board() -> None:
    #first row
    print('\n'*2)
    print("\t\t   |   |   ")
    print("\t\t {b1} | {b2} | {b3}  ".format(b1=game_board[0],b2=game_board[1],b3=game_board[2]))
    print("\t\t___|___|___")
    # second row
    print("\t\t   |   |   ")
    print("\t\t {b4} | {b5} | {b6} ".format(b4=game_board[3], b5=game_board[4],b6=game_board[5]))
    print("\t\t___|___|___")
    # third row
    print("\t\t   |   |   ")
    print("\t\t {b7} | {b8} | {b9} ".format(b7=game_board[6], b8=game_board[7],b9=game_board[8]))
    print("\t\t   |   |   ")
    print('\n'*2)
```

```python
def display_score_board():
    print("\t\t======================================")
    print("\t\t\t Player1 Score: {}".format(game_score['player1']))
    print("\t\t\t Player2 Score: {}".format(game_score['player2']))
    print("\t\t======================================") 
```

```python
def add_player_token(player, other_player, token, other_player_token):
    # edge case: Reset player tokens when called
    player_tokens[player.lower()] = ''
    player_tokens[other_player.lower()] = ''

    if token.lower() == 'x':
        player_tokens[player] = token.upper()
        player_tokens[other_player] = other_player_token.upper()
        
    else:
        player_tokens[player] = token.upper()
        player_tokens[other_player] = other_player_token.upper()
  
```

```python
# Generate user tokens
def select_player_token(player) -> None:
    while True:
        token = input("Select a token, (X,O): ")
        if token.isalpha() == False or (token.isalpha() and token.lower() not in 'xo'):
            print("Invalid Token, Select X or O!")
            continue
        else:
            # Set the player token
            second_player = 'player2' if player == 'player1' else 'player1' 
            second_player_token = 'x' if token.lower() == 'o' else 'o' 
            add_player_token(player=player, other_player=second_player,token=token, other_player_token=second_player_token)
            
            # print the player token
            print('\n')
            print("============================================")
            print("{} Selected: {}.".format(player.capitalize() ,token.upper()))
            print("Therefore, {}'s token is {}.".format(second_player.capitalize(), second_player_token.upper()))
            print("============================================")
            print('\n')
            break
            
```

```python
# Get Players guesses
def guess_random_num(player) -> int:
    while True:
        num = input("Guess any number between (0,10): ")
        if num.isdigit() == False:
            print('Invalid! Only numbers allowed!')
            continue
        elif int(num) < 0 or int(num) > 10:
            print('Invalid! Guess any number between (0,10)!')
            continue
        else:
            print("{} guessed: {}".format(player, int(num)))
            return int(num)
        
```

```python
# determine who starts the game
def determine_starter(player1_guess, player2_guess):
    # Edge Case => reset game_status when this is called
    game_starter['Player1'] = False
    game_starter['Player2'] = False
    
    player1_range = abs(random_num - player1_guess)
    player2_range = abs(random_num - player2_guess)
    # the player closer to +/- 3 of the number gets to start 
    # if player1 or player 2 guess right
    if player1_guess == random_num:
        game_starter['Player1']  = True

    elif player2_guess == random_num:
        game_starter['Player2'] = True
    # if both players are within range => return the closer player
    elif player1_range <= 3 and player2_range <= 3:
        if player1_range < player2_range:
            game_starter['Player1'] = True
        else:
            game_starter['Player2'] = True
    # else, if one of the player is in range => return that player 
    elif player1_range <= 3:
        game_starter['Player1'] = True
    elif player2_range <= 3:
        game_starter['Player2'] = True
    
    # display the starter
    print("====================================================")
    print("Random number: {} | Player1: {} | Player2: {}".format(random_num, player1_guess,player2_guess))
    winner = 'Player1' if game_starter['Player1'] == True else 'Player2'
    print("{} gets to start!".format(winner))
    global current_player
    current_player = winner.lower()
    print("====================================================")
```

```python
def update_player(player):
    # Alternative player turns based on the current player
    global current_player
    current_player = 'player2' if player.lower() == 'player1' else 'player1'
```

```python
def place_token(player) -> None:
    print("{}'s turn:".format(player))
    while True:
        index = input("{}: Place Token in an number between (0, 8)".format(player.capitalize()))
        if index.isdigit() == False:
            print("Invalid, Only numbers are allowed!")
            continue
        elif int(index) < 0 or int(index) > 8:
            print("Invalid, Only numbers between (0,8) are allowed!")
            continue
        else:
            index = int(index)
            # place token on the board
            player_token = player_tokens[player.lower()]
            # edge case => if select index is occupied
            if game_board[index] != ' ':
                print('Position {} is occupied. Select any other number between (0,8)!'.format(index))
                continue
            elif game_board[index] == ' ':
                # add add player to board
                game_board[index] = player_token
                # display the board after a token placement
                display_board()
                # update the current player of the game
                update_player(player)
                # increment placement counter for the current player
                token_counter[player.lower()] += 1
                break 
            else:
                 # edge case => When are spots are occupied 
                 print('All spots are occupied!\n Determining game status...\n')
                 break
```

```python
def check_win():
    status = 'no-win'
    win_group = ''
    # if there is a match found => return true and print the player that won!
    # Loop through all the indices that result in a win
    for group in winning_comb:
        # edge case => if group contains only ' ' or one of the members of the group has ' ' skip that group
        if game_board[int(group[0])] == game_board[int(group[1])] == game_board[int(group[2])] == ' ':
            continue   
        else:
             # if there is a group with the same index: that player has won.
            if game_board[int(group[0])] == game_board[int(group[1])] and game_board[int(group[1])] == game_board[int(group[2])]:
                status = 'win'
                win_group = group
                break
    
    return (status, win_group)
```

```python
# game status
def game_status():
    # edge case => When all tokens have been on the board and no match => return 'draw' and print game
    status = 'no-win'
    win_group = ''
  
    if token_counter['player1'] + token_counter['player2'] == 9:
        # edge case => if the board is full => check there are not matches on the board
        stat, group = check_win()
        if stat == 'win':
            status = stat
            win_group = group
        else:
            status = 'draw'
    else:
        # check for a win
        stat, group = check_win()
        if stat == 'win':
            status = stat
            win_group = group
        
    if status == 'no-win': 
        print("No winner yet! Game still On!\n")
    elif status == 'draw':
        print('Game over!')
        print('Game ended as a DRAW\n')
    elif status == 'win':
        # else, no one has won => 'no winner'
        winner = 'player1' if player_tokens['player1'] == game_board[int(win_group[0])] else 'player2'
        # update the score board
        game_score[winner] += 1
        # print the winner of the game    
        print("{} won the game!\n".format(winner.capitalize()))
        # display the score board
        display_score_board()
        # set the current player to the winner
        global current_player
        current_player = winner

    return status
```

```python
# Restart the game
def restart_game(current_player):
    status = True
    while True:
        other_player = 'player2' if current_player == 'player1' else 'player1'

        reply1 = input("{}: Would you like to continue? Enter (Y/N).".format(current_player.capitalize()))
        if reply1.isdigit() or reply1.lower() not in  'yn':
            print('Invalid, Enter Y/N only!')
            continue 
        else:
            reply1 = reply1.lower()
            if reply1 == 'n':
                print("{} has quit!\nOnly Two players can play the game!".format(current_player.capitalize()))
                status = False
                break

        # update the current_player 
        
        reply2 = input("{}: would you like to continue? Enter (Y/N).".format(other_player.capitalize()))
        if reply2.isdigit() or reply1.lower() not in  'yn':
            print('Invalid, Enter Y/N only!')
            continue 
        else:
            reply2 = reply2.lower()
            if reply1 == 'n':
                print("{} has quit!\nOnly Two players can play the game!".format(other_player.capitalize()))
                status = False
            else:
                print("Game continues....")
                break
    if status == True:
        print("{} will go first now!\nSince {} won the last game!".format(current_player, current_player))
    return status
```

```python
# Play game function
def play_game(counter):
    # print the board
    display_board()
    # print score board 
    display_score_board()

    if counter == 0:
        # prompt users for tokens => default player is player1
        select_player_token(player=current_player)
        print("Next, we will determine who gets to start to make a game fair.\nEach player gets to guess a random number between(0,10)\nThe winner starts the game!\n")

        # determine who goes first 
        guess_player1 = guess_random_num(player=current_player)
        guess_player2 = guess_random_num(player='player2')
        determine_starter(player1_guess=guess_player1, player2_guess=guess_player2)

    while True:
        # Next, Token placement 
        place_token(player=current_player)
        # Check game status after each placement
        if game_status() == 'no-win':
            continue
        else:
            break
```

```python
def reset_game():
    # reset the board after each game turn
    global game_board
    game_board = [' ' for num in range(9)]
    # update the token count
    global token_counter
    token_counter['player1'] = 0
    token_counter['player2'] = 0
```

```python
def main_game():
    game_status = True
    game_count = 0
    while game_status:
        play_game(counter=game_count)
        game_count += 1
        game_status = restart_game(current_player=current_player)
        reset_game()
        
```

```python
if __name__ == "__main__":
    welcome_message()
    main_game()
```
